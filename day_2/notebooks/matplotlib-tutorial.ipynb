```python
import numpy as np
from matplotlib import style
from matplotlib import pyplot as plt
```

## Plotting a Basic Line Plot

```python
plt.plot([1, 2, 3], [4, 5, 1], '-')
plt.title('This is my first Plot!')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.show()
```

## Adding Style To Your Graph

```python
style.use('default')
x1, y1 = [5, 8, 10], [12, 16, 6]
x2, y2 = [6, 9, 11], [6, 15, 7]
plt.plot(x1, y1, 'g', label='plot 1')
plt.plot(x2, y2, 'b', label='plot 2')

plt.title('Stylish Graph')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')

plt.legend()
plt.grid(True)
plt.show()
```

## Bar Graphs

```python
style.use('default') # other options are 'classic', 'ggplot', 'default'

plt.bar([2 * x + 1 for x in range(5)], [5, 2, 7, 8, 2], label='Example 1')
plt.bar([2 * x for x in range(1, 6)], [8, 6, 2, 5, 6], label='Example 2')

plt.legend()
plt.xlabel('X-Axis')
plt.ylabel('Quantity')
plt.title('My First Bar Graph')
plt.grid(False)

plt.show()
```

## Histograms

```python
# creating a random sample of population ages
population_ages = np.random.randint(0, 130, (1, 100000)).ravel().tolist()
population_ages
```

```python
# creating the age bins
bins = [x for x in range(0, 131, 10)]
print(bins)
```

```python
# creaing the histogram plot
plt.hist(population_ages, bins, histtype='bar', rwidth=0.8)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Population Age Density')
plt.show()
```

## Stack Plots / Area Plots

```python
# creating a stack plot to track daily activities
days = [x for x in range(1, 6)]
days
```

```python
sleeping = np.array([7, 8, 6, 11, 7])
eating = [2, 3, 4, 3, 2]
working = [7, 8, 7, 2, 2]
playing = [8, 5, 7, 8, 13]
```

```python
style.use('default')

plt.plot(days, sleeping, color='b', label='Sleeping')
plt.plot(days, sleeping + eating, color='m', label='Eating')
plt.plot(days, sleeping + eating + working, color='g', label='Working')
plt.plot(days, sleeping + eating + working + playing, color='r', label='Playing')

plt.stackplot(days, sleeping, eating, working, playing, colors=['b', 'm', 'g', 'r'])
plt.xlabel('Day')
plt.ylabel('Hours')
plt.title('Daily Activities')
plt.legend()
plt.show()
```

## Pie Chart

```python
slices = [7, 2, 2, 13]
activities = ['Sleeping', 'Eating', 'Working', 'Playing']
plt.pie(slices, labels=activities, autopct='%1.1f%%', startangle=90, explode=(.1, 0, 0, 0))
plt.title('Daily Activities')
plt.show()
```

## Multiple Plots

Defining a function  $e^{-t} \cos{2 \pi t}$

```python
def f(t):
    return np.exp(-t) * np.cos(2 * np.pi * t)
```

```python
# creating linspaces
t1 = np.arange(0, 5, 0.1)
t1
```

```python
t2 = np.arange(0, 5, 0.02)
t2
```

```python
# creating subplots inside the plot
plt.subplot(2, 1, 1)
plt.plot(t1, f(t1), 'o')
plt.plot(t2, f(t2))
```

```python
plt.subplot(2, 1, 2)
plt.plot(t2, np.cos(2 * np.pi * t2))
```
